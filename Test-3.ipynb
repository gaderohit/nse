```python
# Generating Heiken Ashi Charts
def HA(test):
    test['HA_Close']=(test['Open']+ test['High']+ test['Low']+ test['Close'])/4
    test.loc[len(test)-1,'HA_Open'] = (test.loc[len(test)-1,'Open']+test.loc[len(test)-1,'Close'])/2
    for i in range(len(test)-2,0,-1):
        test.loc[i,'HA_Open'] = (test.loc[i+1,'HA_Open'] + test.loc[i+1,'HA_Close'])/2
    test['HA_High'] = test[ ['HA_Open','HA_Close','High'] ].max()
    test['HA_Low'] = test[ ['HA_Open','HA_Close','High'] ].min()
    return test
```
